# Modeling

```python
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression

RANDOM_STATE = 6
```

## Creating model and accuracy functions

```python
# Preprocessing pipelines for numerical and categorical data
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Defining models
models = {
    "Linear Regression": LinearRegression(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "Random Forest": RandomForestRegressor(random_state= 16)
}
```

```python
# Function to check model accuracy
def check_model_accuracy(y_test, predictions):
    mse = mean_squared_error(y_test, predictions)
    rmse = root_mean_squared_error(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    
    print(f'Mean Squared Error (MSE): {mse}')
    print(f'Root Mean Squared Error (RMSE): {rmse}')
    print(f'Mean Absolute Error (MAE): {mae}')
    print(f'R-squared (R2): {r2}')
    print("------------------------------------------------------------------------")
    
    return mse, rmse, mae, r2
```

### 1. By removing nan values

```python
# Making database for modelling
data = df.select_dtypes(include=['float64', 'int64', 'int32']).copy()
data['type'] = df['type'].copy()

# Select relevant features
features = list(data.columns)
target = features.pop(features.index('mag'))

# Drop rows with missing target values
data = data.dropna(subset=[target])

# Dropping nan value rows
data.dropna(axis = 0, inplace= True)
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
# Define numerical and categorical features
numeric_features = list(data.select_dtypes(include=['float64', 'int64', 'int32']).columns)
numeric_features.pop(numeric_features.index('mag'))
categorical_features = list(data.select_dtypes(include=['object']).columns)

print(numeric_features)
print(categorical_features)
```

```python
# Column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
# Split the data into training and testing sets
X = data[features]
y = data[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= RANDOM_STATE)
```

```python
# Dictionary to store results
results = {}

# Train and evaluate each model
for name, model in models.items():
    print(name)
    
    # Define the model pipeline
    model_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])
    
    # Train the model
    model_pipeline.fit(X_train, y_train)
    # Make predictions
    predictions = model_pipeline.predict(X_test)
    
    # Evaluate the model
    mse, rmse, mae, r2 = check_model_accuracy(y_test, predictions)
    results[name] = {"MSE": mse, "RMSE": rmse, "MAE": mae, "R2": r2}
    
    # # Residual Plot
    # plt.figure(figsize=(10, 6))
    # plt.scatter(predictions, predictions - y_test, alpha=0.3)
    # plt.hlines(y=0, xmin=predictions.min(), xmax=predictions.max(), colors='red')
    # plt.xlabel('Predicted Values')
    # plt.ylabel('Residuals')
    # plt.title(f'{name} Residual Plot')
    # plt.show()
    
    # # Actual vs Predicted Plot
    # plt.figure(figsize=(10, 6))
    # plt.scatter(y_test, predictions, alpha=0.3)
    # plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
    # plt.xlabel('Actual Values')
    # plt.ylabel('Predicted Values')
    # plt.title(f'{name} Actual vs Predicted')
    # plt.show()
    
    # # Error Distribution Plot
    # plt.figure(figsize=(10, 6))
    # sns.histplot(y_test - predictions, kde=True)
    # plt.xlabel('Prediction Error')
    # plt.title(f'{name} Error Distribution')
    # plt.show()

# # Optionally, print or store the results dictionary
# print("Model Performance Summary:")
# for name, metrics in results.items():
#     print(f"{name}: {metrics}")
```
